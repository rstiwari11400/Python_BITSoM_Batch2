## 4. Numbers & Arithmetic


```python
a=7
b=3
print('Addition')
print(a + b)
print(type(a + b))

print('Subtraction')
print(a - b)
print(type(a - b))

print('Multiplication')
print(a * b)
print(type(a * b))

print('Division')
print(a / b)
print(type(a / b))
print('Floor Division')
print(a // b)
print(type(a // b))

print('Modulus')
print(a % b)
print(type(a % b))

print('Exponentiation')
print(2 ** 5)
print(type(2 ** 5))

```

# String slicing

```python
sentence = 'This is a short sentence about Python.'
# get first character
print(sentence[0])
# get last character
print(sentence[-1])
# get first four characters
print(sentence[:4])
# get characters from index 5 to 10
print(sentence[5:11])

# get alternate characters
print(sentence[::2])

print(sentence[::-1])

A=list(sentence)
print(A)
```

## 6. Strings — Transform & Clean


```python
text = '  PyThOn Rocks  '
print(text.upper())
print(text.lower())
print(text.title())
print(text.swapcase())
```

```python
messy = '  spaced  '
print(messy.strip())
print(messy.strip(' '))
date = '2024-11-11'
print(date.replace('-', '/'))
```

```python
colors = 'red,green,blue'
print(colors.split(','))
sentence = 'Split on spaces'
print(sentence.split())
parts = ['red','green','blue']
print(' - '.join(parts))
```

## 7. Strings — Search & Inspect


```python
message = 'hello python helloz'
print(message.find('python'))
print(message.find('java'))
print(message.find('hello'))
print(message.find('z'))
print(message.count('hello'))
```

```python
filename = 'report.csv'
print(filename.startswith('rep'))
print(filename.endswith('.csv'))
print('abc123'.isalnum())
print('abc1 23'.isalnum())
print('abc-123'.isalnum())
print('Python'.isalpha(), '123'.isdigit())
```

```python
email = 'user@example.com'
print(email.partition('@'))
print(email.split('.'))
poem = 'line1\nline2'
print(poem.splitlines())
invoice = '421'
print(invoice.zfill(6))
```

## 8. Strings — Formatting


```python
template = 'Hello, {}. You have {} new messages.'
print(template.format('Ada', 5))
print('Hello, {name}. Count={count}'.format(name='Ada', count=5))
```

```python
import math
name = 'Ada'
count = 5
print(f'Hello, {name}. You have {count} new messages.')
print(f'pi to 3dp: {math.pi:.3f}')
```

## 9. Lists — Basics

## 10. Lists — Text Workflows


```python
sentence = 'This is a short sentence about Python.'
words = sentence.lower()
words = words.replace('.', '')
words = words.split()
print(words)
unique = sorted(words)
print(unique)
```

```python
sentence = 'This is a short sentence about Python.'
words = sentence.lower().replace('.', '').split()
print(words)
unique = sorted(words)
print(unique)
```
