## Classic Coding Problems
### *Most Frequently Asked on HackerRank, LeetCode, and Campus Interviews*

> These problems test your ability to combine the basics — loops, conditionals, and string/number manipulation — into a complete algorithm.

### CQ1 — Even or Odd *(Beginner — #1 Most Asked)*
Write a program that checks whether a given number is **Even** or **Odd**.

**Expected Output (n = 7):**
```
7 is Odd
```
**Expected Output (n = 12):**
```
12 is Even
```

> 💡 **Hint:** A number is even if `n % 2 == 0` (no remainder when divided by 2). Use a simple `if/else`.

### CQ2 — Swap Two Variables *(Beginner)*
Swap the values of `a = 5` and `b = 10` **without using a third variable**.

**Expected Output:**
```
Before swap: a=5, b=10
After swap:  a=10, b=5
```

> 💡 **Hint:** Python supports tuple unpacking: `a, b = b, a` does the swap in one line. This is the Python-specific trick — no temp variable needed.

### CQ3 — Factorial of a Number *(Beginner — Classic Loop Question)*
Calculate the factorial of `n = 5`.
Factorial: `5! = 5 × 4 × 3 × 2 × 1 = 120`

**Expected Output (n = 5):**
```
Factorial of 5 = 120
```
**Expected Output (n = 0):**
```
Factorial of 0 = 1
```

> 💡 **Hint:** Start with `result = 1`. Loop from 1 to n (inclusive) using `range(1, n+1)` and multiply: `result *= i`. Note: `0! = 1` by definition.

### CQ4 — Palindrome Check *(Most Asked — #1 String Question)*
A **palindrome** reads the same forwards and backwards (e.g., `"racecar"`, `"madam"`).
Write a program to check if a given string is a palindrome.

**Expected Output (word = "racecar"):**
```
racecar is a Palindrome
```
**Expected Output (word = "python"):**
```
python is NOT a Palindrome
```

> 💡 **Hint:** Reverse the string using `word[::-1]`. If `word == word[::-1]`, it's a palindrome. Always convert to lowercase first to handle case differences: `word.lower()`.

### CQ5 — Fibonacci Series *(Most Asked — Classic Loop + Accumulator)*
Print the first **n** numbers of the Fibonacci series.
Fibonacci: each number is the sum of the previous two → `0, 1, 1, 2, 3, 5, 8, 13...`

**Expected Output (n = 8):**
```
0 1 1 2 3 5 8 13
```

> 💡 **Hint:** Start with `a, b = 0, 1`. In each loop iteration: print `a`, then update `a, b = b, a + b`. This shifts both values forward by one position.

### CQ6 — Prime Number Check *(Popular — Needs Loop + Break)*
A **prime number** is only divisible by 1 and itself (e.g., 2, 3, 5, 7, 11...).
Write a program to check if a given number is prime.

**Expected Output (n = 7):**
```
7 is Prime
```
**Expected Output (n = 9):**
```
9 is NOT Prime
```

> 💡 **Hint:** Loop from 2 to `n-1`. If any number divides `n` evenly (`n % i == 0`), it's not prime — break and flag it. Numbers less than 2 are never prime. Efficiency tip: only loop up to `int(n**0.5)` instead of `n-1`.

### CQ7 — Reverse a Number *(Popular — Number Manipulation)*
Reverse the digits of a number. For example, `1234` becomes `4321`.

**Expected Output (n = 1234):**
```
Reversed: 4321
```
**Expected Output (n = 5000):**
```
Reversed: 5
```

> 💡 **Hint:** Convert the number to a string, reverse it with `[::-1]`, and convert back to int: `int(str(n)[::-1])`. The `int()` at the end automatically drops leading zeros (e.g., `0005` → `5`).

### CQ8 — Sum of Digits *(Popular — Loop + Modulo)*
Find the sum of all digits in a number. For example, `1234` → `1+2+3+4 = 10`.

**Expected Output (n = 1234):**
```
Sum of digits: 10
```
**Expected Output (n = 9875):**
```
Sum of digits: 29
```

> 💡 **Hint (Two approaches):**
> 1. **String approach:** Convert to string, loop through each character, convert back to int and accumulate: `sum(int(d) for d in str(n))`.
> 2. **Math approach:** Use `n % 10` to get the last digit, then `n //= 10` to remove it. Repeat until `n == 0`.

### CQ9 — Armstrong Number *(Classic — Exponents + Digits)*
An **Armstrong number** (narcissistic number) equals the sum of its own digits each raised to the power of the number of digits.
Example: `153 = 1³ + 5³ + 3³ = 1 + 125 + 27 = 153` ✅

**Expected Output (n = 153):**
```
153 is an Armstrong number
```
**Expected Output (n = 123):**
```
123 is NOT an Armstrong number
```

> 💡 **Hint:** Step 1 — find `power = len(str(n))` (number of digits). Step 2 — sum each digit raised to that power: `total = sum(int(d)**power for d in str(n))`. Step 3 — check if `total == n`.

### CQ10 — Count Vowels in a String *(Popular String Question)*
Count how many vowels (`a, e, i, o, u`) are in a given string.

**Expected Output (text = "Hello World"):**
```
Vowels found: 3
```
**Expected Output (text = "Python Programming"):**
```
Vowels found: 4
```

> 💡 **Hint:** Define `vowels = "aeiou"`. Loop through each character of the string (converted to lowercase). If the character is `in vowels`, increment a counter.

### CQ11 — Print Multiplication Table *(Beginner — Very Frequently Set as Assignment)*
Print the multiplication table for a given number up to 10.

**Expected Output (n = 5):**
```
5 x 1  = 5
5 x 2  = 10
5 x 3  = 15
...
5 x 10 = 50
```

> 💡 **Hint:** Use `for i in range(1, 11):` and print `f"{n} x {i} = {n*i}"` inside the loop. You can align columns neatly using f-string width: `{i:<2}` (left-align in 2 characters).

### CQ12 — Largest of Three Numbers *(Beginner — Conditional Logic)*
Given three numbers `a = 45`, `b = 72`, `c = 31`, find and print the largest.

**Expected Output:**
```
Largest number is: 72
```

> 💡 **Hint (Two approaches):**
> 1. **Built-in:** `max(a, b, c)` — Python's `max()` works on multiple arguments too.
> 2. **Manual:** Use nested `if/elif/else` to compare all three — good practice for exam questions.

### CQ13 — Anagram Check *(Popular String Question)*
Two strings are **anagrams** if they contain the same characters in any order.
Example: `"listen"` and `"silent"` are anagrams.

**Expected Output (s1="listen", s2="silent"):**
```
listen and silent are Anagrams
```
**Expected Output (s1="hello", s2="world"):**
```
hello and world are NOT Anagrams
```

> 💡 **Hint:** Sort both strings and compare: `sorted(s1.lower()) == sorted(s2.lower())`. `sorted()` on a string returns a list of characters in alphabetical order — anagrams will produce identical sorted lists.

### CQ14 — Power of a Number (Without `**`) *(Loop Practice)*
Calculate `base ^ exponent` using a loop (without the `**` operator).
For example, `2^8 = 256`.

**Expected Output (base=2, exp=8):**
```
2 ^ 8 = 256
```
**Expected Output (base=3, exp=4):**
```
3 ^ 4 = 81
```

> 💡 **Hint:** Start with `result = 1`. Loop `exp` times and multiply: `result *= base` on each iteration. This is the same as multiplying base by itself exponent times.

### CQ15 — FizzBuzz *(The #1 Most Asked Coding Interview Question Worldwide)*
Print numbers from 1 to 20 with these rules:
- If divisible by **both** 3 and 5 → print `"FizzBuzz"`
- If divisible by **3 only** → print `"Fizz"`
- If divisible by **5 only** → print `"Buzz"`
- Otherwise → print the number

**Expected Output (first 15 lines):**
```
1
2
Fizz
4
Buzz
Fizz
7
8
Fizz
Buzz
11
Fizz
13
14
FizzBuzz
...
```

> 💡 **Hint:** ⚠️ The order of your `if/elif` checks matters! Always check the **combined** condition (`n % 3 == 0 and n % 5 == 0`) FIRST. If you check `n % 3 == 0` first, `FizzBuzz` will never print because `15 % 3 == 0` is true and it stops there.
