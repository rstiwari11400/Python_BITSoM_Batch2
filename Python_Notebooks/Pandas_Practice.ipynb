## Section 5 — GroupBy (Split-Apply-Combine)

GroupBy lets you answer questions like:  
*"What is the average Math score per city?"*

The pattern is always:
1. **Split** → group rows by a column (e.g., City)
2. **Apply** → compute something (mean, max, count…)
3. **Combine** → get the result as a new table

```python
df.groupby("City")["Math"].mean()
#          ^^^^^^  ^^^^^^  ^^^^^^
#          group   column   agg
```

## Section 7 — Map and Apply

| `df.apply(func, axis=1)` | Each row | Combine multiple columns into a new value |
